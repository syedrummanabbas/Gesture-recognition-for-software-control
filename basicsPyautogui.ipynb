```python
import pyautogui
```

### Mouse Functions


```python
# Print  the resolution of your screen
print(pyautogui.size())
```

```python
# print the current positon of the mouse
print(pyautogui.position())
```

```python
# Moves the mouse over time (3 sec)
pyautogui.moveTo(100,100,3)
```

```python
# Move the mouse relative to its current position (3 sec)
pyautogui.moveRel(100,100,3)
```

```python
pyautogui.leftClick(400,400, 2, 3)
```

- Scroll

```python
# Scroll up
pyautogui.scroll(800)
```

```python
# Scroll down
pyautogui.scroll(-800)
```

- Mouse up and down

```python
import time
time.sleep(3)
pyautogui.mouseDown(400,500,button = "left")
pyautogui.moveTo(800,400,3)
pyautogui.mouseUp()
pyautogui.moveTo(1000,400,3)
```

```python
# Spiral drawing using pyautogui
time.sleep(3)
distance = 300
while distance > 0:
    pyautogui.dragRel(distance, 0, 1, button = "left")
    distance = distance - 20
    pyautogui.dragRel(0, distance, 1, button = "left")
    pyautogui.dragRel(-distance, 0, 1, button = "left")
    distance = distance - 20
    pyautogui.dragRel(0, -distance, 1, button = "left")
    time.sleep(1)
```
